```python
import pandas as pd
import numpy as np
import ast
```

```python
base = pd.read_csv("data/base_allocations.csv", engine="pyarrow")
base["Token allocation"] = base["Token allocation"].str.replace(",", "").astype(float)
base.head()
```

```python
votes = pd.read_csv("data/votes.csv", engine="pyarrow")
votes.head()
```

```python
df = votes
df
```

```python
df["base_vote"] = [0 if base["Token allocation"][base['Name']==p].empty else base["Token allocation"][base['Name']==p].to_list()[0] for p in df["project_name"] ]
```

```python
df['votes'] = df['votes'].apply(ast.literal_eval)
```

```python
# np.median(df["votes"][0])
v=[]
for i, row in df.iterrows():
    if row["base_vote"]!=0.:
        minimum = np.inf
        voted = row["base_vote"]
        for vote in row["votes"]:
            if minimum>abs(row["base_vote"]-vote):
                minimum = abs(row["base_vote"]-vote)
                voted = vote
        v.append(voted)
        continue
    v.append(0)
        
```

```python
df["base_vote_round"] = v
```

```python
df.to_csv("data/output.csv", index=False)
```
